```python
#Splitting dependent and independent variables
y=df["Attrition"]
```

```python
y.head()
```

```python
x = df.iloc[:,2:35]
```

```python
x.head()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
x.BusinessTravel = le.fit_transform(x.BusinessTravel)
x.Department = le.fit_transform(x.Department)
x.EducationField = le.fit_transform(x.EducationField)
x.Gender = le.fit_transform(x.Gender)
x.JobRole = le.fit_transform(x.JobRole)
x.MaritalStatus = le.fit_transform(x.MaritalStatus)
x.Over18 = le.fit_transform(x.Over18)
x.OverTime = le.fit_transform(x.OverTime)
```

```python
x
```

```python
x.info()
```

```python
#feature scaling
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
# if run this, it will be in numpy array, so let's convert it to a dataframe x_scaled = ms.fit_tranform(x) 
x_scaled = pd.DataFrame(ms.fit_transform(x), columns=x.columns) 
```

```python
x_scaled
```

```python
#Splitting Data into Train and Test.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.head()
```

## MODEL BUILDING LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test) 
```

```python
y_pred
```

```python
y_test
```

```python
#Accuracy score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
pd.crosstab(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
Roc-AUC curve
```

```python
probability=lr.predict_proba(x_test)[:,1]
probability
```

```python
y
```

```python
y_test_binary = (y_test == 'Yes').astype(int)
fpr,tpr,threshsholds = roc_curve(y_test_binary,probability)
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```
